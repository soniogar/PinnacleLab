## **Import Necessary Libraries:**

```python
from transformers import MarianMTModel, MarianTokenizer
import ipywidgets as widgets
from IPython.display import display
from IPython.display import HTML
```

## **Load the Translation Models:**

```python
# Load models and tokenizers for English to Hindi
model_name_hi = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer_hi = MarianTokenizer.from_pretrained(model_name_hi)
model_hi = MarianMTModel.from_pretrained(model_name_hi)

# Placeholder for Sanskrit model, assuming it's available
# You need to replace this with the correct model if it becomes available
# Mock the Sanskrit model translation function
def translate_to_sanskrit(text):
    # This is a mock function. Replace with actual model translation when available.
    return "This is a mock Sanskrit translation."
model_name_sa = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer_sa = MarianTokenizer.from_pretrained(model_name_sa)
model_sa = MarianMTModel.from_pretrained(model_name_sa)
```

## **Define the Translation Function:**

```python
def translate(text, model, tokenizer):
    tokenized_text = tokenizer(text, return_tensors='pt', padding=True)
    translated_tokens = model.generate(**tokenized_text)
    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    return translated_text
```

## **User Interface:**

```python
# Create text areas for input and output
input_text = widgets.Textarea(
    placeholder='Enter text to translate',
    description='Input:',
    layout=widgets.Layout(width='100%', height='100px')
)

output_text_hi = widgets.Textarea(
    placeholder='Translation ',
    description='Hindi:',
    layout=widgets.Layout(width='100%', height='100px'),
    disabled=True
)

output_text_sa = widgets.Textarea(
    placeholder='Translation',
    description='Sanskrit:',
    layout=widgets.Layout(width='100%', height='100px'),
    disabled=True
)

# Create a button to trigger translation
translate_button = widgets.Button(
    description="Translate",
    button_style='success',  # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to translate',
    icon='language'  # (FontAwesome names without the `fa-` prefix)
)

# Create a header
header = widgets.HTML(
    value="<h1 style='color: #4CAF50;'>English to Hindi and Sanskrit Translator</h1>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

# Create a footer
footer = widgets.HTML(
    value="<h3 style='color: #777;'></h3>",
    layout=widgets.Layout(margin='20px 0 0 0')
)

# Define the button click event handler
def on_translate_button_clicked(b):
    translation_hi = translate(input_text.value, model_hi, tokenizer_hi)
    translation_sa = translate(input_text.value, model_sa, tokenizer_sa)
    output_text_hi.value = translation_hi
    output_text_sa.value = translation_sa

# Attach the event handler to the button
translate_button.on_click(on_translate_button_clicked)

# Arrange widgets in a vertical box
ui = widgets.VBox([header, input_text, translate_button, output_text_hi, output_text_sa, footer])

# Display the widgets
display(ui)
```
